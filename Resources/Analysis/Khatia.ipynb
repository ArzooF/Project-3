```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
global_df = pd.read_sql_query(query, conn)

# Convert the 'Date' column to datetime
global_df['date'] = pd.to_datetime(global_df['date'])

# Extract data for each platform
facebook = global_df['facebook']
myspace = global_df['myspace']

printest = global_df['printest']
stumbleupon = global_df['stumbleupon']
youtube = global_df['youtube']
instagram = global_df['instagram']
reddit = global_df['reddit']

# Line Plot for all platforms
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(global_df["date"], global_df["reddit"], label="reddit")
ax.plot(global_df["date"], global_df["myspace"], label="myspace")
ax.plot(global_df["date"], global_df["instagram"], label="Instagram")
ax.plot(global_df["date"], global_df["stumbleupon"], label="stumbleupon")
ax.plot(global_df["date"], global_df["printest"], label="Printerest")
ax.plot(global_df["date"], global_df["youtube"], label="YouTube")
ax.plot(global_df["date"], global_df["facebook"], label="Facebook")

# Set the x-ticks to show every 20th tick
ax.set_xticks(ax.get_xticks()[::30])



# Rotate the x-ticks for better readability
plt.xticks(rotation=45, size=8)

# Set the labels and title
plt.xlabel("date")
plt.ylabel("Active Users in Millions")
plt.legend()
plt.title("Social Media Statistic Worldwide")

# Show the plot
plt.show()
```

```python
df['date']=pd.to_datetime(df['date'],format='%Y-%m')
df.info()
```

```python
df['Yearly']=df['date'].dt.to_period('Y')
```

```python
feature=df.drop(columns=['date','Yearly'])
feature=feature.columns
yearly=df.groupby('Yearly')[feature].mean().reset_index()
yearly['Yearly']=yearly['Yearly'].dt.to_timestamp()
yearly.info()
```

```python
yearly.head()
```

```python
stats_yearly=yearly.describe()
stats_yearly=pd.DataFrame(stats_yearly).reset_index()
stats_yearly=stats_yearly.drop(columns='Yearly')
stats_yearly
```

```python
stats_yearly=stats_yearly[stats_yearly['index']=='mean']
stats_yearly
```

```python
import numpy as np
transposed_df=np.transpose(stats_yearly)
transposed_df
```

```python
new_df = pd.DataFrame(transposed_df)
new_df = new_df.rename(columns=new_df.iloc[0]).reset_index()
new_df=new_df.drop(0)
new_df
```

```python
threshold = 5.0
new_df.loc[new_df['mean'] < threshold, 'index'] = 'Other'
df_treshold = new_df.groupby('index', as_index=False).sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.pie(df_treshold['mean'], labels=df_treshold['index'], autopct='%1.1f%%', startangle=140)
plt.title('Mean Values of Social Media Platforms US')
plt.axis('equal')
```

```python
top_five=new_df.sort_values(by='mean',ascending=False)
top_five=top_five[:5].reset_index(drop=True)
top_five
```

```python
sns.catplot(data=top_five,x='mean',y='index',kind='bar')
plt.xlabel('percentage of social media popularity')
plt.ylabel('social media')
plt.title('Rangking of Social Media Popularity')
```
